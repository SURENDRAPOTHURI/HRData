**Data Feature**

Our data contains 36 columns, and the primary columns that we believe will provide the necessary information for analysis and achieving our objectives are as follows:

1. Employee_Name: The name of the employee.
2. EmpID: Employee identification number.
3. Salary: The employee's salary.
4. Termd: A binary identifier indicating whether the employee has been terminated (0 or 1).
5. Position: The job position/title of the employee.
6. Department:The department to which the employee is assigned.
7. ManagerName: The name of the employee's manager.
8. RecruitmentSource: The source from which the employee was recruited.
9. PerformanceScore:The performance score of the employee.
10. EngagementSurvey: Employee engagement survey results.
11. EmpSatisfaction: Employee satisfaction level.
12. SpecialProjectsCount: The number of special projects the employee is involved in last 6 months.
13. LastPerformanceReview_Date:The date of the employee's most recent performance review.
14. DaysLateLast30:The number of days the employee was late in the last 30 days.
15. Absences: The number of absences recorded for the employee.

These columns contain crucial data about employees and their performance, making them crucial for various HR analytics and related tasks.

````python
import numpy as np
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dense

# Function to create a Keras model
def create_model(optimizer='adam', hidden_units=[32, 32, 32]):
    model = Sequential()
    
    # Input layer
    model.add(Dense(hidden_units[0], input_dim=X_train_scaled.shape[1], activation='relu'))
    
    # Hidden layers
    for units in hidden_units[1:]:
        model.add(Dense(units, activation='relu'))
    
    # Output layer
    model.add(Dense(1, activation='sigmoid'))
    
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# Create a KerasClassifier for use with GridSearchCV
model = KerasClassifier(build_fn=create_model, epochs=50, batch_size=1, verbose=1)

# Define the hyperparameters and their possible values
param_grid = {
    'optimizer': ['adam', 'sgd', 'rmsprop'],
    'hidden_units': [(16, 16, 16), (32, 32, 32), (64, 64, 64)]
}

# Perform grid search
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
grid_result = grid.fit(X_train_scaled, y_train)

# Summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))

# Evaluate the best model on the test set
best_model = grid_result.best_estimator_
test_loss, test_accuracy = best_model.model.evaluate(X_test_scaled, y_test)
print(f'Test Loss: {test_loss}, Test Accuracy: {test_accuracy}')
````
